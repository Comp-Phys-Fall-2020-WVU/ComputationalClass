As a reminder, remember that when you discuss integrals, you use the Riemann approximation:

$$
\int_a^b f(x) dx = lim_{h \rightarrow 0} \left[ h \sum_{i=1}^{(b-a)/h} f(x_i) \right] \approx \sum_{i=1}^N w_i f(x_i)
$$

## Trapezoid

Okay, I have something I like. Let's compute:

$$
Ax^4 + Bx^3 + Cx^2 + Dx + E
$$

Where $A=-0.1, B=1.0, C=-3.1, D=3.6, E=0.5$. I've just picked nice numbers to give an interesting plot and integral.

```python
x = np.linspace(0, 5.5, 500)
f = np.poly1d([-0.1, 1.0, -3.1, 3.6, 0.5])
y = f(x)
x_i = np.linspace(0, 5.5, 4)
y_i = f(x_i)

fig, ax = plt.subplots()
plt.plot(x, y)
plt.plot(x_i, y_i, "o-r")
plt.plot(x_i, y_i * 0, "o-r")
plt.vlines(x_i, y_i * 0, y_i, "r")
plt.show()
```

Let's try integrating with several different values of N (picking odd values for later use - we could have even numbers if we wanted). Let's also compare with `np.trapz`.

```python
for i in 3, 5, 7, 11, 21, 101, 1001:
    x_i = np.linspace(0, 5.5, i)
    y_i = f(x_i)
    h = np.diff(x_i)[0]
    my_intg = 1 / 2 * h * np.sum(y_i[:-1] + y_i[1:])
    of_intg = np.trapz(y_i, x_i)
    print(f"{i: 5} {my_intg: 16.12f} {of_intg: 16.12f}")
```

We can also calculate the exact value, since it's just a polynomial. In fact, since we built it with a specialized class in numpy, we can just call `.integ` to get the analytic integral (and then we call that with our points):

```python
f.integ()(5.5)
```

### Simpson's rule  (evenly spaced points)

We'll get the `integrate` subpackage from scipy - it has a lot of integration functions. The idea is to approximate the function between ($x_i,x_{i+1}$ with a polynomial of degree 2.

```python
from scipy import integrate
```

Let's make sure we are selecting the times we want with our indexing expressions:

```python
index = list(range(7))
index[:-1:2], index[1::2], index[2::2]
```

Looks good! Now let's try the rule, comparing with `integrate.simps`:

```python
for i in 3, 5, 7, 11, 21, 101, 1001:
    x_i = np.linspace(0, 5.5, i)
    y_i = f(x_i)
    h = np.diff(x_i)[0]
    my_intg = 1 / 3 * h * np.sum(y_i[:-1:2] + 4 * y_i[1::2] + y_i[2::2])
    of_intg = integrate.simps(y_i, x_i)
    print(f"{i: 5} {my_intg: 16.12f} {of_intg: 16.12f}")
```

## Gaussian quadrature

In the Gaussian quadratuve, we use the following approximation
$$
\int_a^b f(x) dx  \approx \sum_{i=1}^N w_i(x_i) f(x_i) 
$$
where now $w_i$ are called quadrature weights. Now, as you have freedom to choose where the $x_i$ and the weights are, now the purpose of the Gaussian quadrature is to find the best values of $x_i$ and $w_i$ which approximate this integral. For that, we start by doing controlled approximations into the function.


Now let us consider a cubic model for our function $\hat{f}(x) = a_0 + a_1 x + a_2 x^2 + a_3 x^3$. Then we want
$$
w_1 f(x_1) + w_2 f(x_2) = \int_{-1}^1 f(x) dx
$$
Why do we use this form? because you have 4 parameters to be found, which express a polynomial of degree 3 (in general for 2*n parameters, we need polynomial of degree 2n-1).

We proceed by trying to find the weights and the $x_i$ positions. Let us say that we want to approximate the result of the integral only two values.

+ Linear Model
$$
c_1 f(x_1) \approx \int_{-1}^{1} f(x) dx
$$
now, in this expression, there are two unknows on the left hand side, meaning that we should estimate the function by $\hat{f}(x) = a_0 + a_1 x$. We need two equations then to find the two unknown. As our estimation is a linear function and the integral is a linear operator, we can apply every term independently. What I mean is that
$$
c_1 a_0 = \int_{-1}^{1} a_0 dx \rightarrow c_1 = 2
$$
the following would be the linear part
$$
c_1 f(x_1) = c_1 x_1 = \int_{-1}^{1} x dx = 0  \rightarrow x_1 = 0
$$
therefore, in this approximation $\int_{-1}^{1} f(x) dx = 2 f(0)$.

+ Cubic model
It happens that quadratures can only be done with odd models. Now, for this case $\hat{f}(x) = a_0 + a_1 x + a_2 x^2 + a_3 x^3$ and for this, we will obtain that 
$$
w_1 f(x_1) + w_2 f(x_2) = \int_{-1}^1 f(x) dx
$$

$$
f(x)=a_0 \;\;\; w_1 a_0 + w_2 a_0 = \int_{-1}^1 a_0 dx = 2 
$$
$$
f(x)=x \;\;\; w_1 x_1 + w_2 x_2 = \int_{-1}^1 x dx = 2 
$$
$$
f(x)=x^2 \;\;\; w_1 x_1^2 + w_2 x_2^2 = \int_{-1}^1 x^2 dx = \frac{2}{3} 
$$
$$
f(x)=x^3 \;\;\; w_1 x_1^3 + w_2 x_2^3 = \int_{-1}^1 x^2 dx = 0 
$$

Now we have 4 equations and 4 unknowns. By doing a bit of algebra $w_1 = w_2 = 1$ and $x_1 = - 1/\sqrt{3}, x_2 =  1/\sqrt{3}$. Therefore $\int_{-1}^1 f(x) dx = f(-\sqrt{3}/3)+f(\sqrt{3}/3)$. Here we are approximating the integral up to any polynomial of order three!!.

There is an exact result based on the **Legendre Polynomials**. These poynomials, $P_n(x)$ are polynomials that increase degree with $n$. These polynomials are tabulated. For example, the first polynomials are $p_0(x)=1, p_1(x) = x, p_2(x) = x^2-1/3, p_3(x)= x^3-3x/5$. Now, the roots of these polynomials happen to be the $x_i$ values we are looking for. Similarly, you can also derive that
$$
w_i = \frac{2}{(1-x_i^2)\left[\frac{dP_n}{dx}(x_i)\right]^2}
$$

Or in a different form:

If $x_1, x_2, ... x_n$ are the roots of the $nth$ Legendre polynomial, then

$$
w_i = \int_{-1}^1 \prod_{j=1, i \neq j}^n \frac{x-x_j}{x_i-x_j} dx
$$

Ok, now that we have a method that changes the evaluation of the integral by approximating it with Legendre polynomials, let see how it works in practice.

```python
p = np.poly1d([7, -8, -3, 3])
```

```python
x = np.linspace(-1, 1)
fig, ax = plt.subplots()
ax.plot(x, p(x))
plt.show()
```

```python
p.integ()(1) - p.integ()(-1)
```

```python
p(-np.sqrt(1 / 3)) + p(np.sqrt(1 / 3))
```

### Making Gaussian quadrature:

Ok, before we implement the quadrature, we need to solve a problem. Until now we have been solving the integral between (-1,1) but let us suppose, we want to integrate is between (a,b). To solve this, we will make a scaling of the variables. Let me define the variable 
$$
u = \frac{2x -b-a}{b-a}
$$
This means that when $x=a, u=-1$ and $x=b, u=1$.
With this, we need to transform the integral
$$
\int_a^b f(x) dx = \frac{b-a}{2} \int_{-1}^{1} f\left(\frac{b-a}{2} u + \frac{a+b}{2} \right) = \frac{b-a}{2} \int_{-1}^{1} h(x) dx
$$
Therefore, the quadrature will be of the form
$$
\int_a^b f(x) dx = \frac{b-a}{2} \int_{-1}^{1} h(x) dx \approx \frac{b-a}{2} \sum_{i=1}^N w_i(x_i) h(x_i) 
$$

Now that we have discussed the limits, now let me discuss some properties of Legendre polynomials which will be useful to us:

+ Recursion relation
$$
P_0(x)=1 \\
P_1(x) =x \\
n P_n(x) = (2n-1) x P_{n-1}(x) - (n-1)P_{n-2}(x)
$$

+ Recursion relation for the Legendre polynomial derivatives

$$
P_n'(x) = \frac{n}{x^2-1} (x P_n(x) - P_{n-1}(x))
$$

+ Finding roots

By an iterative process (Newton-Rapson)
$$
x_0 = \cos \left( \pi \frac{i - 0.25}{n + 1/2} \right) \\
x_{n+1} = x_n - \frac{P_n(x_n)}{P_n'(x_n)}
$$

+ Weights

$$
w_i = \frac{2}{(1-x_i^2) \left[ P_n'(x_i) \right]^2}
$$

Slightly improved version:

```python
max_in = 11

# here we store the weights and the positions on the functions, basically w_i 
w = np.zeros(2001, dtype=np.float64)

x = np.zeros(2001, dtype=np.float64)


def func_e(x):
    return np.exp(-x)


def gauss(npts, job, a, b, x, w, vmin=0.0, vmax=0.0):
    
    m = i = j = t = t1 = pp = p1 = p2 = p3 = 0.0

    eps = 3e-14  # Accuracy: ******ADJUST THIS*******!
    
    # here we have the "integer division"
    
    m = (npts + 1) // 2
    
    for i in range(1, m + 1):
        
        t = np.cos(np.pi * (float(i) - 0.25) / (float(npts) + 0.5))

        t1 = 1
        
        while abs(t - t1) >= eps:
            p1 = 1.0
            p2 = 0.0
            
            for j in range(1, npts + 1):
                p3 = p2
                p2 = p1
                p1 = ((2.0 * float(j) - 1) * t * p2 - (float(j) - 1.0) * p3) / ( float(j))
                
            pp = npts * (t * p1 - p2) / (t * t - 1.0)
            t1 = t
            t = t1 - p1 / pp

        x[i - 1] = -t
        x[npts - i] = t
        w[i - 1] = 2.0 / ((1.0 - t * t) * pp * pp)
        w[npts - i] = w[i - 1]

    if job == 0:
        for i in range(0, npts):
            x[i] = x[i] * (b - a) / 2.0 + (b + a) / 2.0
            w[i] = w[i] * (b - a) / 2.0

    if job == 1:
        for i in range(0, npts):
            xi = x[i]
            x[i] = a * b * (1.0 + xi) / (b + a - (b - a) * xi)
            w[i] = (
                w[i]
                * 2.0
                * a
                * b
                * b
                / ((b + a - (b - a) * xi) * (b + a - (b - a) * xi))
            )

    if job == 2:
        for i in range(0, npts):
            xi = x[i]
            x[i] = (b * xi + b + a + a) / (1.0 - xi)
            w[i] = w[i] * 2.0 * (a + b) / ((1.0 - xi) * (1.0 - xi))


def gaussint(no, min, max):
    quadra = 0.0
    gauss(no, 0, min, max, x, w)
    for n in range(0, no):
        quadra += func_e(x[n]) * w[n]
    return quadra


for i in range(3, max_in + 1, 2):
    result = gaussint(i, 0.0, 1.0)
    print(" i ", i, " err ", abs(result - 1 + 1 / np.exp(1)))
```

Much better version:

```python
def gauss_new(npts, job, a, b, vmin=0.0, vmax=0.0, eps=3e-14):

    x = np.zeros(npts, dtype=np.float64)
    w = np.zeros(npts, dtype=np.float64)

    for i in range((npts + 1) // 2):
        t = np.cos(np.pi * (i + 0.75) / (npts + 0.5))

        t1 = 1
        while abs(t - t1) >= eps:
            p1 = 1.0
            p2 = 0.0
            for j in range(npts):
                p3 = p2
                p2 = p1
                p1 = ((2 * j + 1) * t * p2 - j * p3) / (j + 1)

            pp = npts * (t * p1 - p2) / (t ** 2 - 1)
            t1 = t
            t = t1 - p1 / pp

        x[i] = -t
        x[~i] = t
        print(i,~i)
        w[i] = 2 / ((1 - t ** 2) * pp ** 2)
        w[~i] = w[i]

    if job == 0:
        w = w * (b - a) / 2
        x = x * (b - a) / 2 + (b + a) / 2

    elif job == 1:
        # Use old x for w and x calcs
        w = w * 2 * a * b ** 2 / (b + a - (b - a) * x) ** 2
        x = a * b * (1 + x) / (b + a - (b - a) * x)

    elif job == 2:
        # Use old x for w and x calcs
        w = w * 2 * (a + b) / (1 - x) ** 2
        x = (b * x + b + 2 * a) / (1 - x)

    else:
        raise RuntimeError(f"Job ({job}) must be 0, 1, or 2")

    return x, w


def gaussint_new(f, no, min, max):
    quadra = 0.0
    x, w = gauss_new(no, 0, min, max)
    for n in range(no):
        quadra += f(x[n]) * w[n]
    return quadra


for i in range(3, max_in + 1, 2):
    result = gaussint_new(func_e, i, 0, 1)
    print("i", i, "err", abs(result - 1 + 1 / np.exp(1)))
```
